## Search replies for each tweet using tweet id

```python
a_8=pd.read_csv('hashtags_tweets.csv')
```

```python
def get_replies(tweet):
    conversation_id=tweet[2]
    start=datetime.strptime(tweet[0], '%Y-%m-%d %H:%M:%S')
    end=start+timedelta(days=5)    
    c = twint.Config()
    c.To = tweet[1]
    c.Since = str(start)
    c.Until = str(end)
    c.Replies=True
    c.Pandas=True
    c.Store_object = True
    twint.run.Search(c)
    temp=twint.output.panda.Tweets_df
    l={}
    for ind in temp.index: 
        if int(temp['conversation_id'][ind])==int(conversation_id):
            l[temp['username'][ind]]=temp['tweet'][ind]
    return l
```

```python
from datetime import datetime, timedelta
a_8['replies']=a_8[['date','username','conversation_id']].apply(get_replies, axis=1)
#temp=a_8.apply(lambda row : get_replies(row['date'], row['username'], row['conversation_id']), axis = 1)
```

```python
import re
def clean(tweet):
    tweet=' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split())
    return tweet
```

```python
df['clean_tweet']=df['tweet'].apply(clean)
```
